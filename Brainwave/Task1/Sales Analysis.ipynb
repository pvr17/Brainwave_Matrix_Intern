## Order and Ship Date Analysis

```python
df2.sample(4)
```

```python
# Convert Order Date and Ship Date to datetime format
df2['Order Date'] = pd.to_datetime(df2['Order Date'], format='%d/%m/%Y')
df2['Ship Date'] = pd.to_datetime(df2['Ship Date'], format='%d/%m/%Y')
```

```python
# Calculate the time difference in days
df2['Shipping Time (Days)'] = (df2['Ship Date'] - df2['Order Date']).dt.days
```

```python
# Compute the average shipping time
average_shipping_time = df2['Shipping Time (Days)'].mean()
print(f"The average shipping time is {average_shipping_time:.2f} days.")
```

```python
# Plot the distribution of shipping times
plt.figure(figsize=(10, 6))
plt.hist(df2['Shipping Time (Days)'], bins=range(df2['Shipping Time (Days)'].min(), df2['Shipping Time (Days)'].max() + 1), edgecolor='black')
plt.title('Distribution of Shipping Times')
plt.xlabel('Shipping Time (Days)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Analyze patterns and outliers
shipping_time_summary = df2['Shipping Time (Days)'].describe()
print(shipping_time_summary)

# Identify potential outliers (e.g., shipping times significantly higher than the mean)
outliers = df2[df2['Shipping Time (Days)'] > df2['Shipping Time (Days)'].mean() + 3 * df2['Shipping Time (Days)'].std()]
print(f"Number of outliers: {len(outliers)}")
```

## Top Customers by Sales:

```python
# Group by customer name and sum the sales
customer_sales = df2.groupby('Customer Name')['Sales'].sum()
```

```python
# sort the customer by total sales in descending order
top_customers = customer_sales.sort_values(ascending = False).head(10)
```

```python
print("Top 10 Customers by Total Sales:")
print(top_customers)
```

```python
top_10_sales_contribution = top_customers.sum()
```

```python
overall_sales = df2['Sales'].sum()
```

```python
contibution_percentage = (top_10_sales_contribution / overall_sales) * 100
```

```python
print(f"\nTotal sales contribution by the top 10 customers: ${top_10_sales_contribution:.2f}")
print(f"Percentage contribution to overall sales: {contibution_percentage:.2f}%")
```
